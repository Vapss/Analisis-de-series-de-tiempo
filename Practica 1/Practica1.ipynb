```python
# Convert Dates

#df['Dates'] = pd.to_datetime(df['Dates'], format='%Y-%m-%d')

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d%H:%M:%S')
```

```python
# Group data from 2012 to 2015

start_date = '01-01-2012'
end_date = '01-01-2015'

mask = (df['Date'] > start_date) & (df['Date'] <= end_date)
df = df.loc[mask]
df = df.replace(',', '.', regex=True)

# Create DF of Dates and time


df
```

```python
DatesTimeDF = df[['Dates','Time']]

NoDateTime = df.drop(['Dates','Time'], axis=1)
cols1 = NoDateTime.columns

NoDateTime[cols1] = NoDateTime[cols1].apply(pd.to_numeric, errors='coerce')
```

```python
# By 15 minutes interval and normalize data by dividing by 4

norm15min = NoDateTime.iloc[:,0:372]

norm15min.dtypes

# Append Date and Time to norm15min at the beginning of the DF

norm15min = norm15min/4
```

```python
norm15min['Dates'] = DatesTimeDF['Dates']
norm15min['Time'] = DatesTimeDF['Time']

norm15min
```

```python
# Graficar en una sola imagen el consumo medio de todos los clientes desde el 2012-01-01 hasta el 2015-01-01 en lapsos de 15 minutos

norm15min.groupby([pd.Grouper(key='Time', freq='15min')]).mean().plot()
df.groupby([pd.Grouper(key='Time', freq='15min')]).mean().plot()
```

```python
# By month interval

df.groupby([pd.Grouper(key='Dates', freq='1M')]).sum().plot()
```

```python
# Mean by day with std deviation

df.groupby([pd.Grouper(key='Dates', freq='1D')]).mean().plot()
```
